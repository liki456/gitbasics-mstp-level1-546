```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
```

```python
#Creating dataset
import pandas as pd
boston_df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df["land price"]=boston.target
```

```python
boston_df.head()
```

```python
x=boston_df.drop("land price",axis=1)
x
```

```python
x.values
```

```python
y=boston_df["land price"]
y
```

```python
y.values
```

```python
#Ridge Regression =sum of error+sum of squares of magnitude of coefficient (alpha)
#coefficient (alpha)
from sklearn.linear_model import Ridge

```

```python
rr=Ridge(alpha=1,solver='svd')
```

```python
rr
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
rr.fit(x_train,y_train)
#print(x_train)
```

```python
y_predi=rr.predict(x_train)
```

```python
y_predi
```

```python
#score
rr.score(x_train,y_train)
```

```python
rr.score(x_test,y_test)
```

```python
rr.coef_
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lrtrain_score=lr.score(x_train,y_train)
print("LR Train score",lrtrain_score)
```

```python
lr.coef_
```

```python
import matplotlib.pyplot as plt
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lr.coef_,marker=".",label="Linear coefficient",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Rigde vs Linear")
plt.legend()
plt.show()
```

```python
rr1=Ridge(alpha=100,solver="cholesky")
```

```python
rr1
```

```python
rr1.fit(x_train,y_train)
```

```python
rr1.coef_
```

```python
plt.plot(rr1.coef_,'go')
plt.plot(lr.coef_,'y*')
plt.plot(rr.coef_,marker="*",label="Ridge coefficient",color="r")
plt.plot(lr.coef_,marker=".",label="Linear coefficient",color="y")
plt.xlabel("coefficient index")
plt.ylabel("coefficient magnitude")
plt.title("Rigde vs Linear")
plt.legend()
plt.show()
```
