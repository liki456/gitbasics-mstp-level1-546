```python
# Representing a sin wave

import numpy as np
x=np.arange(1,10,0.1)
y=np.sin(x)
pl.title('sine wave a=sin(wt)')
pl.xlabel('time')
pl.ylabel('amplitude')
pl.plot(x,y)
pl.grid()
pl.show()

```

```python
yrs=df.columns[2::]    
income=df.iloc[1,2::]
pl.plot(yrs,income)
pl.xlabel('years')
pl.ylabel('alaska_income')
pl.title('income from 2005 to 2013')
pl.grid()
pl.show()
```

```python
yrs=df.columns[-5::]    
income=df.iloc[0:,2:]
pl.plot(yrs,income)
pl.xlabel('years')
pl.ylabel('incomes')
pl.title('income from 2005 to 2013')
pl.grid()
pl.savefig('all states income.jpg')
pl.show()
```
